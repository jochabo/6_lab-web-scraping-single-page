### 1. Scrape the current top 100 songs and their respective artists, and put the information into a pandas dataframe.

```python
# 1. import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 2. find url and store it in a variable, we have the first 100 songs on the first page!
url = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
# 3. download html with a get request
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# 4.2. check that the html code looks like it should
soup
```

```python
#get first title from google chrome
soup.select('#chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p > cite')
```

```python
#list with all the titles
soup.select('cite.title')
```

```python
#get first artist from google chrome
soup.select('#chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p > em')
```

```python
#get artists
soup.select('em.artist')
```

```python
#initialize empty lists
title = []
artist = []

# define the number of iterations of our for loop
# by checking how many elements are in the retrieved result set
# (this is equivalent but more robust than just explicitly defining 250 iterations)
num_iter = len(soup.select('cite.title'))

titlelist = soup.select('cite.title')
artistlist = soup.select('em.artist')

# iterate through the result set and retrieve all the data
for i in range(num_iter):
    title.append(titlelist[i].get_text())
    artist.append(artistlist[i].get_text())

```

```python
# each list becomes a column
music = pd.DataFrame({"title":title,
                       "artist":artist
                      })
music.head()
```
